```python
def extract_primary(df):
    email_columns = ["Email(default)","Email(Personal)","Email(Work)"]
    phone_columns = ["Phone(default)","Phone(Work)","Phone(Home)","Phone(Mobile)","Phone(Main)","Phone(Other)"]
    Email_primary= []
    phone_primary= []
    Cols_to_delete = ["Email(default)","Email(Personal)","Email(Work)", 
                  "Phone(default)","Phone(Work)","Phone(Home)","Phone(Mobile)","Phone(Main)","Phone(Other)"]
    
    for e_cols in df[email_columns]:
        for item in df[e_cols]:
            if pd.notna(item):
                Email_primary.append(item)
                
    for phn_cols in df[phone_columns]:
        for item in df[phn_cols]:
            if pd.notna(item):
                phone_primary.append(item)
    
    df["Email (primary)"] = Email_primary
    df["Phone (primary)"] = phone_primary
    df["Country"] = df["Country"].replace("US","United States")
    df["Country"] = df["Country"].replace("UK","United Kingdom")
    df = df.drop(Cols_to_delete,axis=1)
    df = df[["First Name","Last Name","Company","Email (primary)","Phone (primary)","Country","Zip Code"]]
    
    return df
```

```python
cleaned_sheet = extract_primary(df)
```
